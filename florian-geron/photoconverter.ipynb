```python
import os
import shutil
from PIL import Image
import pillow_heif

# Register HEIC support with Pillow
pillow_heif.register_heif_opener()

input_folder = "/Users/floriangeron/Desktop/ToConvert"
output_folder = "/Users/floriangeron/Desktop/Converted"

# Function to process a folder recursively
def process_folder(src_folder, dst_folder):
    os.makedirs(dst_folder, exist_ok=True)

    for filename in os.listdir(src_folder):
        src_path = os.path.join(src_folder, filename)
        dst_path = os.path.join(dst_folder, filename)

        if os.path.isdir(src_path):
            # Recurse into subfolder
            process_folder(src_path, dst_path)
            continue

        name, ext = os.path.splitext(filename)
        ext = ext.lower()

        # Copy JPG/JPEG directly
        if ext in [".jpg", ".jpeg"]:
            shutil.copy2(src_path, dst_path)
            print(f"Copied JPG: {src_path} → {dst_path}")
            continue

        # Convert other image formats to JPG
        try:
            with Image.open(src_path) as img:
                # Convert any image to RGB (avoids transparency issues)
                if img.mode != "RGB":
                    img = img.convert("RGB")

                output_jpg_path = os.path.join(dst_folder, name + ".jpg")
                img.save(output_jpg_path, "JPEG")
                print(f"Converted: {src_path} → {output_jpg_path}")

        except Exception as e:
            print(f"Skipping {src_path}: {e}")

# Run the processing
process_folder(input_folder, output_folder)
```
